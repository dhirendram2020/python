```python
#Hashtable-Name	Mean	Sd	Zscore	Tot-Bucket-Usages	Outlier-Factor-Ratio	Max-buckect	Used-Bucket	Total-element	
#                                                                         Min	Max	Used -Bucket-Ratio

import pandas as pd
import numpy as np
df = pd.read_csv('all_in_one.csv', usecols=['Hashtable-Name', 'Bucket-Usages(%)', 'Outlier-Factor-Ratio', 'Max-buckect', 'Used-Bucket', 'Total-element', 'Min', 'Max', 'Used -Bucket-Ratio'])
df['Bucket-Usages(%)'] = df['Bucket-Usages(%)'] * 100
df['Log-OFR'] = np.log(df['Outlier-Factor-Ratio'])
df
```

```python
#Sort data frame based on 'Outlier-Factor-Ratio' and print unique records for each hash table which has maximum value of 'Outlier-Factor-Ratio' 
sdf = df.sort_values(by='Log-OFR', ascending = (False)).drop_duplicates(['Hashtable-Name'])
sdf
```

```python
import matplotlib.pyplot as plt
ax = sdf.plot(kind='bar',x='Hashtable-Name',y='Log-OFR', width=1, figsize=(10, 8), zorder=1, title='Max Outlier-Factor-Ratio of each Hash Table')
ax.set_xlabel("Hashtable-Name")
#ax.ylim=(0,3)
ax.set_ylabel("Log-OFR")
ax.axhline(y=3,xmin=0, xmax=1, color='r', linestyle='--', lw=2)
```
