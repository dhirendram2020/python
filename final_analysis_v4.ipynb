```python
#Hashtable-Name	Mean	Sd	Zscore	Tot-Bucket-Usages	Outlier-Factor-Ratio	Max-buckect	Used-Bucket	Total-element	
#                                                                         Min	Max	Used -Bucket-Ratio	Min-Zscore	Max-Zscore

import pandas as pd 
df = pd.read_csv('all_in_one.csv', usecols=['Hashtable-Name', 'Bucket-Usages(%)', 'Outlier-Factor-Ratio', 'Max-buckect', 'Used-Bucket', 'Total-element', 'Min', 'Max', 'Used -Bucket-Ratio', 'Min-Zscore', 'Max-Zscore'])
df['Bucket-Usages(%)'] = df['Bucket-Usages(%)'] * 100
df
```

```python
sdf = df.sort_values(by='Max-Zscore', ascending = (False)).drop_duplicates(['Hashtable-Name'])
sdf


```

```python
import matplotlib.pyplot as plt
ax = sdf.plot(kind='bar',x='Hashtable-Name',y='Max-Zscore', width=1, figsize=(10, 8), zorder=1, title='Max Zscore Values of Various Hash Table')
ax.set_xlabel("Hashtable-Name")
#ax.ylim=(0,3)
ax.set_ylabel("Max-Zscore")
ax.axhline(y=3,xmin=0, xmax=1, color='r', linestyle='--', lw=2)
```
